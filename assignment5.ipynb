```python
!pip -q install -U transformers datasets evaluate sentencepiece

import time
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

from datasets import load_dataset
import evaluate
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

dataset = load_dataset("xsum")
N_TEST = 12
test = dataset["test"].shuffle(seed=42).select(range(N_TEST))
articles = test["document"]
references = test["summary"]

model_list = [
    "sshleifer/distilbart-cnn-12-6",
    "t5-small",
    "facebook/bart-base",
    "google/pegasus-xsum",
    "sshleifer/distilbart-xsum-12-6"
]

rouge = evaluate.load("rouge")

def count_params(model):
    return sum(p.numel() for p in model.parameters())

@torch.inference_mode()
def generate_summaries(model, tokenizer, texts, batch_size=2, max_input_len=256, max_new_tokens=40, min_new_tokens=8):
    outs = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        enc = tokenizer(
            batch,
            truncation=True,
            padding=True,
            max_length=max_input_len,
            return_tensors="pt"
        ).to(device)
        gen = model.generate(
            **enc,
            max_new_tokens=max_new_tokens,
            min_new_tokens=min_new_tokens,
            num_beams=2,
            length_penalty=1.0,
            early_stopping=True
        )
        outs.extend(tokenizer.batch_decode(gen, skip_special_tokens=True))
    return outs

def run_eval(model_name, articles, references):
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)
    params_m = count_params(model) / 1e6

    _ = generate_summaries(model, tokenizer, articles[:2], batch_size=1)

    t0 = time.perf_counter()
    preds = generate_summaries(model, tokenizer, articles, batch_size=2)
    t1 = time.perf_counter()

    latency_ms = ((t1 - t0) / len(articles)) * 1000.0

    scores = rouge.compute(predictions=preds, references=references, use_stemmer=True)
    r1 = float(scores["rouge1"])
    rL = float(scores["rougeL"])

    return {
        "model": model_name,
        "rouge1": r1,
        "rougeL": rL,
        "latency_ms_per_sample": latency_ms,
        "params_million": params_m
    }

start = time.perf_counter()
results = []
for m in model_list:
    print("Evaluating:", m)
    results.append(run_eval(m, articles, references))
end = time.perf_counter()
print("Total runtime (sec):", round(end - start, 2))

df = pd.DataFrame(results)

criteria_cols = ["rouge1", "rougeL", "latency_ms_per_sample", "params_million"]
X = df[criteria_cols].to_numpy(dtype=float)
norm = np.sqrt((X**2).sum(axis=0))
Xn = X / (norm + 1e-12)

weights = np.array([0.40, 0.30, 0.20, 0.10], dtype=float)
Xw = Xn * weights

benefit = np.array([True, True, False, False])
ideal_best = np.where(benefit, Xw.max(axis=0), Xw.min(axis=0))
ideal_worst = np.where(benefit, Xw.min(axis=0), Xw.max(axis=0))

d_best = np.sqrt(((Xw - ideal_best) ** 2).sum(axis=1))
d_worst = np.sqrt(((Xw - ideal_worst) ** 2).sum(axis=1))
topsis_score = d_worst / (d_best + d_worst + 1e-12)

df["topsis_score"] = topsis_score
df["rank"] = df["topsis_score"].rank(ascending=False, method="dense").astype(int)
df_topsis = df.sort_values("topsis_score", ascending=False).reset_index(drop=True)

print(df_topsis[["rank","model","rouge1","rougeL","latency_ms_per_sample","params_million","topsis_score"]])

plt.figure()
plt.barh(df_topsis["model"], df_topsis["topsis_score"])
plt.xlabel("TOPSIS Score (higher = better)")
plt.title("TOPSIS Ranking")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

out_path = "topsis_summarization_xsum_small_results.csv"
df_topsis.to_csv(out_path, index=False)
print("Saved:", out_path)

best = df_topsis.iloc[0]
print("BEST MODEL:", best["model"])
print("TOPSIS Score:", best["topsis_score"])
print("ROUGE-1:", best["rouge1"])
print("ROUGE-L:", best["rougeL"])
print("Latency(ms/sample):", best["latency_ms_per_sample"])
print("Params(M):", best["params_million"])
```
